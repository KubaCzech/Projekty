```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
class PIDAgent():

  def __init__(self, P = 0.005, I = 0.0001, D = 0.4):

    self._num_actions = 2

    # Containers you may find useful.
    self._state = None
    self._action = None
    self._next_state = None

    self.integral = 0
    self.derivative = 0
    self.prev_error = 0

    self.P = P
    self.I = I
    self.D = D


  def select_action(self, observation):
    error = observation[2]
    self.integral += error
    self.derivative = error - self.prev_error
    self.prev_error = error
    pid = self.P * error + self.I * self.integral + self.D * self.derivative
    action = sigmoid(pid)
    action = np.round(action).astype(np.int32)
    return action

  def observe_first(self, timestep):
    # Set current state.
    self._state = timestep.observation

  def observe(self, action, next_timestep):
    # Unpacking the timestep to lighten notation.
    s = self._state
    a = action
    # r = next_timestep.reward
    # g = next_timestep.discount
    next_s = next_timestep.observation

    next_a = self.select_action(next_s)

    self._action = a
    self._next_state = next_s

  def update(self):
    # Optional unpacking to lighten notation.
    s = self._state
    a = self._action

    self._state = self._next_state

#Szwabe: Take a look at https://ekursy.put.poznan.pl/pluginfile.php/1941523/mod_resource/content/1/L2_The_basic_concepts_and_vocabulary_of_reinforcement_learning.pdf
#What type of an intelligent agent is PIDAgent?
```

```python
def CartPole_exam(nr_of_episodes, agent):
  env = gym.make('CartPole-v1')
  episode_results = []
  reward_history = []
  error_history = []

  for episode_nr in range(nr_of_episodes):
    observation, info = env.reset()
    observation = observation.tolist()
    cumulative_reward=0
    time = 0
    finished = False

    while (time < 500 and not finished):
      action = agent.select_action(observation)
      observation, reward, finished, info, _ = env.step(action)

      angle = observation[2]
      error_history.append(angle)
      reward_history.append(reward)

      cumulative_reward += reward
      time+=1

    episode_results.append(cumulative_reward)
  utility = np.mean(episode_results)
  return error_history, utility
```

```python
def objective_function(x):
    P, I, D = x
    tuned_agent = PIDAgent(P, I, D)
    error_history, utility = CartPole_exam(100, tuned_agent)
    error= np.mean(np.array(error_history)**2)
    return error
```
