## Presenting functionality

```python
mixin = FillerNA(categorical_features=['Embarked', 'Cabin'], numerical_features=['Age'])
mixin.fit_transform(data).head(10)
```

```python
mixin = DeckExtractor()
mixin.fit_transform(data).head(10)
```

```python
mixin = TitleExtractor()
mixin.fit_transform(data).head(10)
```

```python
mixin = FamilyTransformer()
mixin.fit_transform(data).head(10)
```

```python
mixin = StandardizeTransformer(features=['Fare', 'Age'])
temp_data = mixin.fit_transform(data)

temp_data.head(10)
```

```python
mixin = TitleExtractor()
temp_data = mixin.fit_transform(data)
mixin = DummiesTransformer()
mixin.fit_transform(temp_data).head(10)
```

```python
mixin = DeckExtractor()
temp_data = mixin.fit_transform(data)

mixin = OrdinalTransformer()
mixin.fit_transform(temp_data).head(10)
```

```python
Pipeline([
    ('Deck', DeckExtractor()),
    ('Filling na', FillerNA(categorical_features=['Embarked'], numerical_features=['Age'])),
    ('Standardization', StandardizeTransformer()),
    ('Family', FamilyTransformer()),
    ('Title', TitleExtractor()),
    ('Dummy', DummiesTransformer()),
    ('Ordinal', OrdinalTransformer()),
    ('Dropper', DropperTransformer())
]).fit_transform(data).head(10)
```
